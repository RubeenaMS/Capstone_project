```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 1) BAR CHART

# Load data and set style
df = pd.read_csv("../data/cleaned_data.csv")
sns.set_theme(style="whitegrid")

# Calculate churn proportions per contract type
contract_churn = (
    df.groupby("Contract")["Churn"].mean().reset_index().sort_values(by="Churn")
)

# Plot
plt.figure(figsize=(9, 5))
sns.barplot(x="Contract", y="Churn", data=contract_churn, palette="Reds")
plt.title("Customer Churn Proportions by Contract Type", fontweight="bold", pad=15)
plt.xlabel("Contract Structure")
plt.ylabel("Churn Rate (Proportion)")
plt.ylim(0, 0.25)
plt.tight_layout()
sns.despine()

# Save and show
plt.savefig("churn_by_contract.png", dpi=300)
plt.show()
```

```python
# 2) BOX PLOT
plt.figure(figsize=(9, 5))
sns.boxplot(x="Churn", y="Tenure", data=df, palette="Set2")
plt.title("Account Lifespan (Tenure) Distribution by Churn Status", fontweight="bold", pad=15)
plt.xlabel("Churn Category (0 = Retained, 1 = Churned)")
plt.ylabel("Tenure Duration (Months)")
plt.tight_layout()
sns.despine()
# Save and show
plt.savefig("tenure_by_churn.png", dpi=300)
plt.show()
```

```python
# 3) HISTOGRAM

plt.figure(figsize=(9, 5))
sns.histplot(
    data=df, x="MonthlyCharges", hue="Churn", multiple="stack", bins=15, palette="viridis"
)
plt.title("Distribution of Monthly Charges Across Customer Profiles", fontweight="bold", pad=15)
plt.xlabel("Monthly Service Fees ($)")
plt.ylabel("Total Account Volume")
plt.tight_layout()
sns.despine()
# Save and show
plt.savefig("monthly_charges_distribution.png", dpi=300)
plt.show()
```

```python
# 4) HEATMAP
numerical_features = ["Tenure", "MonthlyCharges", "TotalCharges", "SeniorCitizen", "Churn"]
correlation_matrix = df[numerical_features].corr()

# Plot
plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, cbar=True
)
plt.title("Correlation Matrix of Numeric Customer Features", fontweight="bold", pad=15)
plt.tight_layout()
sns.despine()
# Save and show
plt.savefig("correlation_heatmap.png", dpi=300)
plt.show()
```

```python
#5) SCATTER PLOT

plt.figure(figsize=(9, 5))
sns.scatterplot(
    data=df, x="Tenure", y="TotalCharges", hue="Churn", style="Churn", palette="magma", alpha=0.8
)
plt.title("Total Charges Accumulated vs. Customer Tenure", fontweight="bold", pad=15)
plt.xlabel("Tenure Duration (Months)")
plt.ylabel("Total Financial Revenue ($)")
plt.tight_layout()
sns.despine()
plt.savefig("total_charges_vs_tenure.png", dpi=300)
plt.show()
```
