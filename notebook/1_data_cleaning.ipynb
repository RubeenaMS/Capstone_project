```python
import pandas as pd


# 1. Import dataset
file_path='../data/raw_data.csv'
df = pd.read_csv(file_path)

# 2. Clean column headers (remove accidental spaces)
df.columns = df.columns.str.strip()

# 3. Clean specific text columns (remove trailing/leading spaces from categories)
df["Contract"] = df["Contract"].str.strip()
df["PaymentMethod"] = df["PaymentMethod"].str.strip()
df["PaperlessBilling"] = df["PaperlessBilling"].str.strip()

# 4. Remove duplicate rows if any exist
df = df.drop_duplicates()

# 5. Save the final output to the data folder
# Note: Ensure you have manually created the empty "data" folder first!
df.to_csv("../data/cleaned_data.csv", index=False)

print(" Cleaned dataset successfully saved to data/cleaned_data.csv!")
print(f" Final structure: {df.shape[0]} rows, {df.shape[1]} columns.")
print(df.head())
```
