```python
# Runs your  statistical & regression tests
import pandas as pd
import scipy.stats as stats
import statsmodels.formula.api as smf

df = pd.read_csv("../data/cleaned_data.csv")

# PART 1: STATISTICAL TEST
# Purpose: Check if pricing differences between the two groups are real

retained = df[df["Churn"] == 0]["MonthlyCharges"]
churned = df[df["Churn"] == 1]["MonthlyCharges"]

t_stat, p_val = stats.ttest_ind(churned, retained, equal_var=False)

print("=== STATISTICAL TEST RESULTS ===")
print(f"Average Monthly Charge (Retained): ${retained.mean():.2f}")
print(f"Average Monthly Charge (Churned):  ${churned.mean():.2f}")
print(f"P-Value: {p_val:.5f}")
if p_val <= 0.05:
    print("Result: Statistically Significant (Price heavily impacts churn)\n")
else:
    print("Result: Not Significant (Price differences are random chance)\n")

# PART 2: REGRESSION TEST
# Purpose: Predict Churn probability using multiple features at once

print("=== REGRESSION TEST RESULTS ===")
model = smf.logit("Churn ~ Tenure + MonthlyCharges + C(Contract)", data=df).fit()
print(model.summary())
```
